```python
#Split X_train into two dataframes - one with only numerical variables and the other one with only categorical variables.
df2_numerical = df2.select_dtypes(include='float64')
df2_categorical = df2.select_dtypes(include='object')
```

```python
#Look for NAs in the 2 dataframes before imputation
print(df2_numerical.isnull().sum().sum())
print(df2_categorical.isnull().sum().sum())
```

```python
#For numerical columns in df2 replace by mean of each column all NAs in that column
df2_numerical.fillna(df2_numerical.mean(), inplace=True)
```

```python
#Check that you have no NAs in your numerical dataframes anymore
print(df2_numerical.isnull().sum().sum())
```

```python
#Load required package
import numpy as np
```

```python
#Constant impute for categorical variables
from sklearn.impute import SimpleImputer
# Creating an imputer object that looks for 'Nan' values, then replaces them with missing_value for categorical variables
miss_constant_imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent', copy=False)
df2_categorical = pd.DataFrame(miss_constant_imputer.fit_transform(df2_categorical),index = df2_categorical.index.values, columns=df2_categorical.columns)
```

```python
#Check number of NAs in df2_categorical after imputation
print(df2_categorical.isnull().sum().sum())
```

NORMALIZE THE NUMERICAL VARIABLES

```python
from sklearn.preprocessing import StandardScaler
```

```python
#Normalize X_train_numerical
sc = StandardScaler()
sc.fit(df2_numerical)
df2_numerical = pd.DataFrame(sc.transform(df2_numerical), index = df2_numerical.index, columns = df2_numerical.columns)
```

```python
df2_numerical.head(5)
```

RECONCILE CATEGORICAL AND NUMERICAL DATAFRAMES

```python
#Now that imputation has been successfully made for both numerical and categorical variables, 
#Merge back df2_categorical and df2_numerical and rename it to df2_new. 
df2_new = pd.concat([df2_numerical, df2_categorical], axis=1)
```

```python
#Make sure you have no null values in both datasets after concatenate
print(df2_new.isnull().sum().sum())
```
